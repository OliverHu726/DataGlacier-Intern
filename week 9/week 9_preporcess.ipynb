```python
# Method 1: Imputing by mode
def replace_na_mode(column):
    mode_val = column.mode().compute().iloc[0]
    return column.fillna(mode_val)

mode_columns = ['fecha_alta', 'ind_empleado', 'indext', 'indfall', 'indresi', 'pais_residencia', 'sexo']
meta = {'fecha_alta':'object',
        'ind_empleado': 'object',
        'indext':'object',
        'indfall':'object',
        'indresi':'object',
        'pais_residencia':'object',
        'sexo':'object'}
df[mode_columns] = df[mode_columns].map_partitions(replace_na_mode, meta=meta)
```

```python
# Method 2: Imputing by 'Null'
def replace_na_null(column):
    return column.fillna('Null')

mode_columns = ['conyuemp', 'canal_entrada', 'cod_prov', 'nomprov', 'renta', 'segmento']
meta = {'conyuemp':'object',
        'canal_entrada': 'object',
        'cod_prov':'object',
        'nomprov':'object',
        'renta':'object',
        'segmento':'object'}

df[mode_columns] = df[mode_columns].map_partitions(replace_na_mode, meta=meta)
```

```python
# Method 3: Imputing by creating a new variable
def create_rencent_primary(ult_fec_cli_1t):
    return ult_fec_cli_1t.map(lambda x: '1' if pd.notna(x) else '0')

df['recent_primary'] = df['ult_fec_cli_1t'].map_partitions(create_rencent_primary, meta = ('recent_primary', object))
```
