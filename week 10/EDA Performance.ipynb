```python
# NA values
df.select(pl.col("*").is_null().sum())
```

```python
# Fill NA by mode
mode_columns = ['ind_empleado', 'pais_residencia', 'sexo', 'ind_nuevo', 'indrel', 'indrel_1mes', 'tiprel_1mes', 'indresi', 
                'indext', 'indfall', 'tipodom', 'ind_actividad_cliente','ind_nomina_ult1','ind_nom_pens_ult1' ]
for col in mode_columns:
    mode_values = df[col].mode()[0]
    df = df.with_columns(pl.col(col).fill_null(mode_values))
```

```python
# Fill NA by mean
mean_columns = ['age', 'antiguedad', 'renta']
for col in mean_columns:
    df = df.with_columns(pl.col(col).fill_null(strategy = 'mean'))
```

```python
# Fill NA by 'Unknown'
Unknown_columns = ['conyuemp', 'cod_prov', 'segmento', 'canal_entrada']
for col in Unknown_columns:
    df = df.with_columns(pl.col(col).fill_null('Unknow'))
```

```python
# 'indrel_1mes'(MonthBeginCustomerType), replace ['1.0', '2.0', '3.0', '4.0'] by ['1', '2', '3', '4']
def replace_words(category, category_map):
    for k, v in category_map.items():
        category = category.replace(k, v)
    return category
```

```python
mapping = {'1.0': '1',
           '2.0': '2',
           '3.0': '3',
           '4.0': '4'}
df = df.with_columns(
    pl.col("indrel_1mes").map_elements(lambda x: replace_words(x, category_map=mapping)).alias("indrel_1mes")
)
```

```python
def products_bar(product_columns, data, custom_labels=None):
    """
    Plot the number of customers for each product in the specified product_columns.

    Args:
        product_columns (list): List of product column names to analyze.
        data (pd.DataFrame): The Pandas DataFrame containing the data.
        custom_labels (list, optional): Custom labels for the product columns. Default is None.

    Returns:
        None (displays a bar chart).
    """
    # Create a dictionary to store product counts
    product_counts = {}

    # Count the number of customers for each product
    for col in product_columns:
        product_counts[col] = data[col].sum()

    # Convert the dictionary to a Pandas DataFrame
    product_counts_df = pd.DataFrame.from_dict(product_counts, orient='index', columns=['Customer Count'])

    # Plot a bar chart
    ax = product_counts_df.plot(kind='bar', figsize=(10, 6))
    plt.title('Number of Customers for Each Product')
    plt.xlabel('Product')
    plt.ylabel('Customer Count')

    # Set custom labels if provided
    if custom_labels:
        ax.set_xticklabels(custom_labels, rotation=30)

    plt.show()
```

```python
# Short, Mid, and Long term deposites:
product_columns = ['ind_deco_fin_ult1', 'ind_deme_fin_ult1', 'ind_dela_fin_ult1']
custom_labels = ['Short-term-deposit', 'Mid-term-deposit', 'Long-term-deposites']
products_bar(product_columns, df, custom_labels)
```

```python
# Accounts:
  # Saving Account, Derivada Account, Payroll Account, Junior Account, 
  # Más particular Account, particular Account, particular Plus Account, e-account, Home Account
product_columns = ['ind_ahor_fin_ult1', 'ind_cder_fin_ult1', 'ind_cno_fin_ult1', 'ind_ctju_fin_ult1',
                  'ind_ctma_fin_ult1', 'ind_ctop_fin_ult1','ind_ctpp_fin_ult1', 'ind_ecue_fin_ult1', 'ind_viv_fin_ult1']
custom_labels = ['Saving Account','Derivada Account', 'Payroll Account', 'Junior Account',
                'Más particular Account', 'particular Account', 'particular Plus Account', 'e-account', 'Home Account']
products_bar(product_columns, df, custom_labels)
```

```python
# Bamk Services: 
 # Guarantees, Funds, Mortgage, Pensions, Loans,
 # Taxes, Credit Card, Securities, Payroll, Pensions, Direct Debit
product_columns = ['ind_aval_fin_ult1', 'ind_fond_fin_ult1', 'ind_hip_fin_ult1', 'ind_plan_fin_ult1', 'ind_pres_fin_ult1',
                  'ind_reca_fin_ult1', 'ind_tjcr_fin_ult1', 'ind_valo_fin_ult1', 'ind_nomina_ult1', 'ind_nom_pens_ult1', 'ind_recibo_ult1']
custom_labels = ['Guarantees', 'Funds', 'Mortgage', 'Pensions', 'Loans', 'Taxes', 'Credit Card', 'Securities', 'Payroll',
                'Pensions', 'Direct Debit']
products_bar(product_columns, df, custom_labels)
```

```python
# 6 Products are selected for furthur EDA:
  # Long_term_diposite, Payroll_account, e_account, and Direct_Debit
  # 'ind_dela_fin_ult1', 'ind_cno_fin_ult1', 'ind_ecue_fin_ult1',  'ind_recibo_ult1'
column_mapping = {'ind_dela_fin_ult1': 'Long_term_diposite',
                  'ind_cno_fin_ult1': 'Payroll_account',
                  'ind_ecue_fin_ult1': 'e_account',
                 'ind_recibo_ult1': 'Direct_Debit'}

df = df.rename(column_mapping)
```

```python
column_mapping = {'ind_empleado': 'Employee_index',
                  'pais_residencia': 'Customer_Country_Residence',
                  'ind_nuevo': 'New_customer',
                  'antiguedad': 'Customer_seniority',
                  'indrel': 'Primary_Customer',
                  'indrel_1mes': 'MonthBeginCustomerType',
                  'tiprel_1mes': 'MonthBeginCustomerRelation',
                  'indresi' : 'Residence_index',
                  'indext': 'Foreigner_index',
                  'conyuemp': 'Spouse_index',
                  'canal_entrada': 'Channel_join',
                  'indfall': 'Decreased_index',
                  'nomprov': 'Province name',
                  'ind_actividad_cliente': 'Active',
                  'renta': 'Gross_income'}

df = df.rename(column_mapping)
```

```python
# One last step before EDA: Assign age group
age_groups = ['Under 18', '18-35', '36-65', 'Over 65']
age_ranges = [(0, 17), (18, 35), (36, 65), (65, df['age'].max())]
def assign_age_group(age):
    for i, (start, end) in enumerate(age_ranges):
        if start <= age <= end:
            return age_groups[i]
    return 'Unknown'

df = df.with_columns(age_group = pl.col('age').map_elements(assign_age_group))
```

```python
def pie_chart(column, data):
    counts = data[column].value_counts()
    plt.pie(counts['counts'], labels=counts[column], autopct='%1.1f%%', startangle = 45)
    plt.title(column + ' Distribution')
    plt.axis('equal')
```

```python
def plot_bar(column, data):
    plot_df = data[column].value_counts().sort(by = 'counts', descending = True).head(10).to_pandas()
    plt.figure(figsize=(10, 6))
    sns.barplot(x = plot_df['counts'], 
                y = plot_df[column], 
                palette = 'pastel', 
                orient = 'h')
    plt.title('Count of ' + column + ' (Ordered)')
    plt.xticks(rotation=45)
    plt.xlabel('counts')
    plt.ylabel('column')
    plt.show()
```

```python
plot_bar('Channel_join', Long_term_diposite)
```

```python
plot_bar('Province name', Long_term_diposite)
```

```python
def first_contract_date_plot(data):
    plot_df = data['fecha_alta'].value_counts().sort(by = 'fecha_alta')
    plt.figure(figsize=(12, 6))
    plt.plot(plot_df['fecha_alta'], plot_df['counts'], marker='.', linestyle='-', color='b')
    plt.title('Long-term-deposite Customer first date contract(any product)')
    plt.xlabel('Date')
    plt.ylabel('Number of customers')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.show()
```

```python
first_contract_date_plot(Long_term_diposite)
```
